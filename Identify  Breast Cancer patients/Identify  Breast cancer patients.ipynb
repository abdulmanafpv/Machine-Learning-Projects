```python
import pandas as pd
bc=pd.read_csv("C:\\Users\\user\\Desktop\\datas\\breast-cancer-wisconsin.data",header=None)
print(bc)
```

```python
X=bc.drop([6,10], axis=1)
print(X)
```

```python
y_raw=bc[10].values
print(y_raw)
```

 We want to change our Y to binary format

```python
from sklearn.preprocessing import LabelEncoder
label_enc=LabelEncoder()
Y=label_enc.fit_transform(y_raw)
print(len(Y))
print(Y)
```

```python
import sklearn.model_selection as ms
Xtrain,Xtest,Ytrain,Ytest=ms.train_test_split(X,Y,test_size=0.3,random_state=20)
from sklearn.linear_model import LogisticRegression
```

```python
pen_val=['l1','l2']
import numpy as np
C_val=2. **np.arange(-5,10,step=2)
print(C_val)
```

```python
grid_s=[{"C":C_val,'penalty':pen_val}]
model=LogisticRegression(solver='liblinear')
from sklearn.model_selection import GridSearchCV

cv_logr=GridSearchCV(estimator=model,param_grid=grid_s,
                     cv=ms.KFold(n_splits=10))

cv_logr.fit(Xtrain,Ytrain)
```

```python
best_c=cv_logr.best_params_['C']
best_penalty=cv_logr.best_params_['penalty']
print('The best parameters are: c={0} and penalty = {1}'.format(best_c,best_penalty))

b_clf=LogisticRegression(C=best_c,penalty=best_penalty,solver='liblinear')
b_clf.fit(Xtrain,Ytrain)
```

```python
predict=b_clf.predict(Xtest)
print(predict)

y_proba=b_clf.predict_proba(Xtest)
print(y_proba)

print(b_clf.score(Xtest,Ytest))
print(b_clf.coef_)
```
